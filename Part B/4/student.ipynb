```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
student_df = pd.read_csv('student.csv')
student_df.head()
```

```python
student_df.describe()
```

```python
student_df.head(5)
```

```python
print("======Data Headers After Dropping Columns==================")
student_df.drop(['test preparation course','lunch'], axis=1, inplace=True)
print(student_df.head(5))
```

```python
print("======Data Headers After Transforming race/ethnicity Column =======")
student_df ['race/ethnicity'] = student_df ['race/ethnicity'].map({
    'group A': 'Asian Students',
    'group B': 'African Students',
    'group C': 'Afro-Asian Students',
    'group D': 'American Students',
    'group E': 'European Students'
})
print(student_df.head(5))
```

```python
print(" Gender Based Visualization of Student Group")
ax = sns.countplot(x = 'race/ethnicity', hue = 'gender',data = student_df)
ax.set(title = 'Gender Based Visualization of Student Group',
       xlabel = 'race', ylabel = 'Total')
plt.show() 
```

```python
print(" Number of Students on the basis of marks obtained")
interval = (0,40,50,60,75)
categories = ['Failed','Second Class','First Class', 'Distinction']

student_df['marks_plot'] = pd.cut(student_df.mathscore, interval, labels = categories)

ax = sns.countplot(x = 'marks_plot',  data = student_df, hue = 'gender', palette = 'Set3')

ax.set(xlabel='Degree', ylabel='Total',
       title="Number of Students on the basis of marks obtained")
plt.show()
```
